```python
print('Welcome to Child Nutrition Calculator')

name = input('Hi User, Can you please tell me your name? ')
age = int(input(f'Great {name}, Now Please enter your age: '))
weight = float(input('Now, please enter your weight in pounds: '))
height = float(input('Lastly, enter your height in inches: '))

bmi = round((weight / height ** 2) * 703, 2)

print(f'Hello {name}, your BMI is {bmi:.2f}')

if bmi < 16:
    print('You are severely underweight.')
elif bmi < 18.5:
    print('You are underweight.')
elif bmi < 25:
    print('You are healthy.')
elif bmi < 30:
    print('You are overweight.')
else:
    print('You are obese. Consider seeking professional advice.')

foods = {
    'Milk': 1,
    'Egg': 1.55,
    'Rice': 1.3,
    'Lentils': 1.13,
    'Vegetables': 0.85,
    'Meat': 1.43
}

calories = {}
total_calories = 0


for food, factor in foods.items():
    quantity = float(input(f'Enter the quantity in grams of {food}: '))
    calories[food] = quantity * factor
    total_calories += calories[food]

recommended_calories = {
    0: 800,
    2: 1400,
    4: 1800
}

if age >= 0 and age < 8:
    recommended_calories = recommended_calories.get(age, None)
    if recommended_calories is not None:
        if total_calories >= recommended_calories:
            print(f'Congratulations, you are well nourished! Your recommended calorie intake is {recommended_calories} and you are consuming {total_calories:.2f}.')
        else:
            print(f'Sorry, you are mal-nourished. Your recommended calorie intake is {recommended_calories} and you are consuming {total_calories:.2f}.')
    else:
        print('Sorry, the recommended calorie intake for this age group is not available.')
else:
    print('Sorry, this program is designed for children only.')
```
